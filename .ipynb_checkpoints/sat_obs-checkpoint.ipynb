```python
import math
import datetime as dt

import xarray as xr

import numpy as np
from numpy.linalg import inv

import scipy as sp
from scipy import sparse
from scipy.sparse.linalg import inv as spinv
import scipy.stats as stats

import pandas as pd

from os import remove, listdir
from os.path import isfile, join

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

from matplotlib import rcParams

import cartopy.crs as ccrs
import cartopy

# import sys
# sys.path.append('/Users/hannahnesser/Documents/Harvard/Research/Python')
# import inv_plot

rcParams['font.family'] = 'serif'
rcParams['font.size'] = 14
colors = ['DarkOliveGreen','Olive', 'YellowGreen', 'Lime', 'Aqua', 'Navy']

%matplotlib inline
```

```python
obs_base = pd.read_csv('Base_Run/sat_obs.gosat.00.m', delim_whitespace=True,
                  header=0)
obs_base['S_OBS'] = np.sqrt(obs_base['S_OBS'])*1e9
obs_base['GOSAT'] = obs_base['GOSAT']*1e9
obs_base['model'] = obs_base['model']*1e9
obs_base['diff'] = obs_base['model'] - obs_base['GOSAT']

obs_base.head()
```

```python
obs_ref = pd.read_csv('Base_Run/sat_obs_bram.gosat.00.m', delim_whitespace=True,
                      header=0)
obs_ref['S_OBS'] = np.sqrt(obs_ref['S_OBS'])*1e9
obs_ref['GOSAT'] = obs_ref['GOSAT']*1e9
obs_ref['model'] = obs_ref['model']*1e9
obs_ref['diff'] = obs_ref['model'] - obs_ref['GOSAT']
obs_ref = obs_ref[(obs_ref['YYYY'] == 2009) &
                  (obs_ref['MM'].isin([4,5,6,7]))]
print(obs_ref.shape)
obs_ref.head()
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,5))
# plt.subplots_adjust(hspace=0.4)
for j, ds in enumerate([obs_base, obs_ref]):
    for i, month in enumerate(obs_base['MM'].unique()):
        df = ds[ds['MM'] == month].copy()
        mu = df['diff'].mean()
        sns.distplot(df['diff'], hist=False, color=colors[i],
                     ax=ax[j])

        ax[j].axvline(mu, c=colors[i])
        ylim = ax[j].get_ylim()
        ax[j].text(mu, 0.047, '%d : %.2f' % (month, mu),
                fontsize=10,
                va='bottom', ha='center', rotation=90)
    ax[j].set_xlim(-40, 40)
    ax[j].set_ylim(0, 0.045)

ax[0].set_title('Base Run', y=1.3, fontsize=14)
# ax[1].set_title('Perturbation Run', y=1.3, fontsize=14)
ax[1].set_title('Reference Run', y=1.3, fontsize=14)
```

```python
fig, ax = plt.subplots(3,1,figsize=(10,17),
                       subplot_kw={'projection' : ccrs.PlateCarree()})

cmap=['plasma', 'plasma', 'seismic']
vlims = [(1650, 1950), (1650, 1950), (-50, 50)]

for i, quant in enumerate(['GOSAT', 'model', 'diff']):
    df = obs_base[(obs_base['MM'] == 7)]
    ax[i].set_title(quant)
    im = ax[i].scatter(df['LON'], df['LAT'], 
                       c=df[quant], cmap=cmap[i],
                       vmin=vlims[i][0], vmax=vlims[i][1],
                       s=2)
    fig.colorbar(im, ax=ax[i])

for axis in ax.flatten():
    axis.set_xlim(-140,-40)
    axis.set_ylim(10,70)
    axis.coastlines(color='grey')
```

```python
lat_bands = np.arange(10, 71, 10)

obs_base['LAT_BAND'] = pd.cut(obs_base['LAT'], lat_bands)
obs_lat_band = obs_base[['LAT_BAND', 'MM', 'GOSAT', 'model', 'diff']].groupby(['MM', 'LAT_BAND']).mean()
obs_lat_band = obs_lat_band.reset_index()

obs_ref['LAT_BAND'] = pd.cut(obs_ref['LAT'], lat_bands)
obs_ref_lat_band = obs_ref[['LAT_BAND', 'MM', 'GOSAT', 'model', 'diff']].groupby(['MM', 'LAT_BAND']).mean()
obs_ref_lat_band = obs_ref_lat_band.reset_index()

# obs_pert['LAT_BAND'] = pd.cut(obs_pert['LAT'], lat_bands)
# obs_pert_lat_band = obs_pert[['LAT_BAND', 'MM', 'GOSAT', 'model', 'diff']].groupby(['MM', 'LAT_BAND']).mean()
# obs_pert_lat_band = obs_pert_lat_band.reset_index()

fig, ax = plt.subplots(1,2,figsize=(14,5))
for j, ds in enumerate([obs_lat_band, obs_ref_lat_band]):
    for i, lb in enumerate(ds['LAT_BAND'].unique()):
        df = ds[ds['LAT_BAND'] == lb]
        ax[j].plot(df['MM'], df['diff'], label=lb, c=colors[i])
#         df_pert = obs_pert_lat_band[obs_pert_lat_band['LAT_BAND'] == lb]
#         ax[j].scatter(df_pert['MM'], df_pert['diff'], c=colors[i])
    ax[j].set_xlabel('Month')
    ax[j].set_xticks(np.arange(4,8))
    ax[j].set_ylabel('Model - Observations')
    ax[j].set_ylim(-10, 15)
    
ax[0].set_title('Base Run', fontsize=13)
ax[1].set_title('Reference Run', fontsize=13)
ax[1].legend()
```

```python
obs_base = pd.read_csv('Perturbation_Runs/0000/sat_obs.gosat.00.m', delim_whitespace=True,
                      header=0)
obs_base['S_OBS'] = np.sqrt(obs_base['S_OBS'])*1e9
obs_base['GOSAT'] = obs_base['GOSAT']*1e9
obs_base['model'] = obs_base['model']*1e9
obs_base['diff'] = obs_base['model'] - obs_base['GOSAT']

obs_pert = {}
for i in [1,2,823]:
    
    op = pd.read_csv('Perturbation_Runs/%s/sat_obs.gosat.00.m' % str(i).zfill(4),
                     delim_whitespace=True,
                      header=0)
    op['S_OBS'] = np.sqrt(op['S_OBS'])*1e9
    op['GOSAT'] = op['GOSAT']*1e9
    op['model'] = op['model']*1e9
    op['diff'] = op['model'] - op['GOSAT']
    op['pert'] = op['model'] - obs_base['model']
    obs_pert[i] = op

clusters = xr.open_dataset('Jacobian_Construction/Clusters_NA.nc')['Clusters']
```

```python
# Load in the totals 
conc_base = xr.open_dataset('Perturbation_Runs/0000/GEOSChem.SpeciesConc.20090701_0000z.nc4')
# clusters = xr.open_dataset('Jacobian_Construction/Clusters_NA.nc')
# cluster = clusters.where(clusters['Clusters']==1, drop=True)['Clusters']

conc = {'base' : conc_base}
for i in [1,2,823]:
    cp = xr.open_dataset('Perturbation_Runs/%s/GEOSChem.SpeciesConc.20090701_0000z.nc4' % str(i).zfill(4))
    conc['pert_%s' % (str(i).zfill(4))] = cp

for key, c in conc.items():
    # Add a pressure level index
    _, idx = np.unique(c.lev, return_inverse=True)
    idx = idx + 1
    idx = idx[::-1]
    c.coords['lev'] = idx
    c['SpeciesConc_CH4'] = c['SpeciesConc_CH4']*1e9
    c = c.drop(['hyam', 'hybm', 'hyai', 'hybi', 'P0', 'AREA'])
    conc[key] = c
```

```python
fig, ax = plt.subplots(2,2,figsize=(13,7),
                       subplot_kw={'projection' : ccrs.PlateCarree()})

base_l0 = conc['base']['SpeciesConc_CH4'].where(conc['base']['lev'] == 1, drop=True)
pert_l0 = conc['pert_0823']['SpeciesConc_CH4'].where(conc['pert_0823']['lev'] == 1, drop=True)
diff_l0 = pert_l0 - base_l0

base_l0.plot(ax=ax[0,0])
ax[0,0].set_title('Base Run')

pert_l0.plot(ax=ax[0,1])
ax[0,1].set_title('Perturbation Run')

diff_l0.plot(ax=ax[1,0], cmap='seismic', center=0)
ax[1,0].set_title('Perturbation - Base Run')

ax[1,1].set_visible(False)

colors=['white', 'white', 'grey', 'white']
for i, axis in enumerate(ax.flatten()):
#     axis.scatter(clusters['lon'].mean(), clusters['lat'].mean(),
#                  s=3, c='blue')
#     axis.set_xlim(-140,-40)
#     axis.set_ylim(10,70)
    clusters.where(clusters==823, drop=True).plot(color='forestgreen', 
                                                    add_colorbar=False,
                                                    ax=axis)
    axis.coastlines(color=colors[i])
#     i += 1

# ax[1,0].set_xlim(-90, -80)
# ax[1,0].set_ylim(9, 20)
    
```
