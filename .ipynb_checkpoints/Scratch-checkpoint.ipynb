```python
# base_conc = xr.open_dataset('../Base_Run/GEOSChem.SpeciesConc.20090701_0000z.nc4')
pert_conc = xr.open_dataset('Perturbation_Runs/GEOSChem.SpeciesConc.20090701_0000z.nc4')
# clusters = xr.open_dataset('../Jacobian_Construction/Clusters_NA.nc')
# cluster = clusters.where(clusters['Clusters']==1, drop=True)['Clusters']

concs = {'pert' : pert_conc}
for key, conc in concs.items():
    # Add a pressure level index
    _, idx = np.unique(conc.lev, return_inverse=True)
    idx = idx + 1
    idx = idx[::-1]
    conc.coords['lev'] = idx
    conc['SpeciesConc_CH4'] = conc['SpeciesConc_CH4']*1e9
    conc = conc.drop(['hyam', 'hybm', 'hyai', 'hybi', 'P0', 'AREA'])
    concs[key] = conc
```

```python
obs = pd.read_csv('Perturbation_Runs/0000/sat_obs.gosat.00.m', delim_whitespace=True,
                  header=0)
```

```python
obs['S_OBS'] = np.sqrt(obs['S_OBS'])*1e9
so = obs['S_OBS']
SO = xr.Dataset({'SO' : (['Nobs'], so)})

SO.attrs['description'] = 'GOSAT error estimate'
SO['SO'].attrs['standard_name'] = 'S_O'
SO['SO'].attrs['units'] = 'ppb'
    
SO.to_netcdf('SO.nc')
```
